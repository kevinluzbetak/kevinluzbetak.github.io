```python
# @title California Housing Analysis
# prompt: generate sample exploratory code to utilize and plot data from /content/sample_data/california_housing_test.csv

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df_housing = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Display basic information about the dataset
print(df_housing.info())
print(df_housing.describe())

# Explore relationships between variables
# Example: Scatter plot of housing median age vs median house value
plt.figure(figsize=(8, 6))
sns.scatterplot(x='housing_median_age', y='median_house_value', data=df_housing)
plt.title('Housing Median Age vs. Median House Value')
plt.xlabel('Housing Median Age')
plt.ylabel('Median House Value')
plt.show()

# Example: Histogram of median income
plt.figure(figsize=(8, 6))
sns.histplot(df_housing['median_income'], bins=20)
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()

# Example: Correlation matrix
correlation_matrix = df_housing.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# You can further explore the data by:
# - Creating box plots to compare median house value across different regions
# - Analyzing the distribution of total rooms or bedrooms
# - Investigating the relationship between population and median house value
# - etc.
```

```python
# prompt: generate sample exploratory code to utilize and plot data from /content/sample_data/california_housing_train.csv

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Import seaborn

# Load the training dataset
df_housing_train = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Display basic information about the training dataset
print(df_housing_train.info())
print(df_housing_train.describe())

# Explore relationships between variables in the training dataset
# Example: Scatter plot of total rooms vs median house value
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_rooms', y='median_house_value', data=df_housing_train)
plt.title('Total Rooms vs. Median House Value (Training Data)')
plt.xlabel('Total Rooms')
plt.ylabel('Median House Value')
plt.show()

# Example: Box plot of median house value by ocean proximity
plt.figure(figsize=(10, 6))

# Check if 'ocean_proximity' column exists, and fix if necessary
if 'ocean_proximity' not in df_housing_train.columns:
    print("Warning: 'ocean_proximity' column not found. Check your data.")
    # If column name has a typo, fix it here:
    # df_housing_train = df_housing_train.rename(columns={'incorrect_column_name': 'ocean_proximity'})
else:
    sns.boxplot(x='ocean_proximity', y='median_house_value', data=df_housing_train)
    plt.title('Median House Value by Ocean Proximity (Training Data)')
    plt.xlabel('Ocean Proximity')
    plt.ylabel('Median House Value')
    plt.xticks(rotation=45)
    plt.show()

# Example: Pairplot for selected features
sns.pairplot(df_housing_train[['median_income', 'median_house_value', 'housing_median_age', 'total_rooms']], diag_kind='kde')
plt.show()

# You can further explore the training data by:
# - Comparing the distribution of features between training and test datasets
# - Creating more complex visualizations to identify patterns
# - etc.
```
